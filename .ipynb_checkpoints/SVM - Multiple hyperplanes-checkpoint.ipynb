```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;sns.set(font_scale=1.2)
from sklearn import svm
```

```python
recipes = pd.read_csv("../Datasets/Cupcakes vs Muffins.csv")
recipes.head(10)
```

```python
# Defining X and Y
labels = np.where(recipes['Type']=='Muffin',0,1)
ingredients = recipes[['Flour','Milk','Sugar']].values
```

```python
model =svm.SVC(kernel='linear')
model.fit(ingredients,labels)
```

```python
def cupcakeOrMuffin(flour,sugar,milk):
    if model.predict([[flour,sugar,milk]])==0:
        print('You\'re having a muffin')
    else:
        print('You\'re having a Cupcake')
```

```python
cupcakeOrMuffin(50,60,20)
```

```python
sns.lmplot(x="Flour",y='Milk', data=recipes, hue ="Type", palette='Set2', fit_reg=False, scatter_kws={'s':70})
```

# Getting the hyperplanes for each 

```python
w = model.coef_[0]
a = -w[0] / w[1]
xx= np.linspace(30,60)
yy = a * xx -(model.intercept_[0]) / w[1]

b = model.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])

b= model.support_vectors_[2]
yy_up = a * xx + (b[1] - a * b[0])
```

```python
sns.lmplot(x="Flour",y='Milk', data=recipes, hue ="Type", palette='Set1', fit_reg=False, scatter_kws={'s':70})
plt.plot(xx,yy, linewidth=2,color='black')
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(xx,yy_down, 'k--',color='red')
plt.plot(xx,yy_up, 'k--',color='blue')
```

```python
w = model.coef_[0]
a = -w[0]/w[2]
xx= np.linspace(30,60)
yy = a*xx-(model.intercept_[0]) / w[2]

b = model.support_vectors_[0]
yy_down = a*xx + (b[1]-a*b[0])


b= model.support_vectors_[3]
yy_up = a*xx + (b[1]-a*b[0])
```

```python
sns.lmplot(x="Flour", y="Sugar", data=recipes, hue="Type", palette="Set1",fit_reg=False,scatter_kws={'s':70})
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(xx,yy_down, 'k--',color='red')
plt.plot(xx,yy_up, 'k--',color='blue')
```

##This shows that, using this model, you cannot base on only sugar or only milk to make predictions


```python
# Now if you put them together somehow
#Using both parameters- Using support vectors from both
w = model.coef_[0]
a = -w[0]/w[1]
xx= np.linspace(30,60)
yy = a*xx-(model.intercept_[0]) / w[1] # Flour and milk

b = model.support_vectors_[0] # flour
yy_down = a*xx + (b[1]-a*b[0]) # Support vectors for flour


b= model.support_vectors_[3]# Sugar
yy_up = a*xx + (b[1]-a*b[0]) #Support vectors for Sugar

sns.lmplot(x="Flour",y='Milk', data=recipes, hue ="Type", palette='Set1', fit_reg=False, scatter_kws={'s':70})
plt.plot(xx,yy, linewidth=2,color='black')
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(xx,yy_down, 'k--',color='red')
plt.plot(xx,yy_up, 'k--',color='blue')
```
